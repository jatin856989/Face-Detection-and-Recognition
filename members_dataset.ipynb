```python
import pickle
import face_recognition
import os
import cv2
```

```python
path = "train_images"
images = []
person_name = []
myList = os.listdir(path)
print(myList)
```

```python
type(myList)
```

```python
for i in myList:
    current_img = cv2.imread(f"{path}/{i}")   # fstring  # path(folder)/imagefile
    images.append(current_img)
    person_name.append(os.path.splitext(i)[0])
print(person_name)
```

```python
def faceEncodings(images):
    encode_List = []
    for img in images:
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        encode = face_recognition.face_encodings(img)[0]
        encode_List.append(encode)
    return encode_List
```

```python
encode_List_known = faceEncodings(images)
```

```python
encode_List_known
```

```python
data = {"encodings": encode_List_known, "names": person_name}
f = open("face_enc", "wb")   # wb opens the file in binary format for writing encodings
f.write(pickle.dumps(data))
f.close()
```
